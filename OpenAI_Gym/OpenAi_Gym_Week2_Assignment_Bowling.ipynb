```
"""
Utility functions to enable video recording of gym environment and displaying it
To enable video, just do "env = wrap_env(env)""
"""

def show_video():
  mp4list = glob.glob('video/*.mp4')
  if len(mp4list) > 0:
    mp4 = mp4list[0]
    video = io.open(mp4, 'r+b').read()
    encoded = base64.b64encode(video)
    ipythondisplay.display(HTML(data='''<video alt="test" autoplay 
                loop controls style="height: 400px;">
                <source src="data:video/mp4;base64,{0}" type="video/mp4" />
             </video>'''.format(encoded.decode('ascii'))))
  else: 
    print("Could not find video")
    

def wrap_env(env):
  env = Monitor(env, './video', force=True)
  return env
```

# Bowling-v4 with Q-Learning

```
alpha = 0.618
class Q_Learner(object):
    def __init__(self, env):
        self.obs_shape = env.observation_space.shape
        self.obs_high = env.observation_space.high
        self.obs_low = env.observation_space.low
        self.action_shape = env.action_space.n
        # Create a multi-dimensional array (aka. Table) to represent the
        # Q-values
        self.Q = np.zeros([210*160, env.action_space.n])
        self.alpha = alpha  # Learning rate

    def get_action(self, episode, obs):
        if random.random() < (0.8 / (episode*.1)): 
            action = random.randint(0,env.action_space.n-1)
        else:
            action = np.argmax(self.Q[obs[:,:,0]])
        return action

    def learn(self, obs, action, reward, next_obs):
        self.Q[obs[:,:,0],action] += self.alpha * (reward + np.max(self.Q[next_obs[:,:,0]]) - self.Q[obs[:,:,0],action]) #3
```

```
MAX_NUM_EPISODES = 101
def train(agent, env):
    best_reward = -float('inf')
    for episode in range(1, MAX_NUM_EPISODES):
        done = False
        obs = env.reset()
        total_reward = 0
        reward = 0
        while not done:
            action = agent.get_action(episode, obs)
            next_obs, reward, done, info = env.step(action)
            agent.learn(obs, action, reward, next_obs)
            obs = next_obs
            total_reward += reward
        if total_reward > best_reward:
            best_reward = total_reward
        print("Episode#:{} reward:{} best_reward:{}".format(episode,
                                     total_reward, best_reward))
```

```
print("Action Space {}".format(env.action_space.n))
print("State Space {}".format(env.observation_space.shape[0]))
```

```
agent = Q_Learner(env)
```

```
train(agent, env)
```

```
def test(agent, env):
    done = False
    obs = env.reset()
    total_reward = 0.0
    while not done:
        env.render()
        action = np.argmax(agent.Q[obs[:,:,0]])
        next_obs, reward, done, info = env.step(action)
        obs = next_obs
        total_reward += reward
    env.close()
    return total_reward
```

```
for _ in range(200):
    test(agent, env)
env.close()
show_video()
```
